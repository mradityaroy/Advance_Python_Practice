```python
import sys

class Person:
    def __init__(self, name):
        self._name = name
        
    def __getattr__(self, attr):             # object.attr RHS use
        print("get: ", attr)
        if attr == 'name':
            return self._name
        else:
            raise AttributeError (attr)
            
    def __setattr__(self, attr, value):       # obj.attr = value LHS use
        print("set: ", attr)
        if attr == 'name':
            attr = '_name'
        self.__dict__[attr] = value           # Avoid recursion
        
    def __delattr__(self, attr):
        print("del: ", attr)
        if attr == 'name':
            attr = '_name'
        del self.__dict__[attr]
        

def main():
    p1 = Person("Rodger")
    print(p1.name)
    
    p1.name = 'Rafa'
    print(p1.name)
    del(p1.name)
    p1.abc = 'rzd'
    
    sys.exit
    
main()
```

# Attributeintercept Demo 7

```python
import sys

# using __getattribute__ instead of __getattr__
# Because we intercept al the attributes,
# __dict__ entries will be visible in demo_7
# those were not visible in demo_6
# This establishes that only unknown attributes are
# caught by __getattr__ and all attributes are caught
# by __getattribute__

class Person:
    def __init__(self, name):
        self._name = name
    def __getattribute__(self, attr):       # on [obj.any]
        print('get: ' + attr)               # Intercept all names
        if attr == 'name':
            attr = '_name'                  # Map to internal name
        return object.__getattribute__(self, attr)     # Avoid looping here

    def __setattr__(self, attr, value):       # obj.attr = value LHS use
        print('set: ', attr)
        if attr == 'name':
            attr = '_name'
        self.__dict__[attr] = value         # Avoid recursion
        
    def __delattr__(self, attr):
        print("del: ", attr)
        if attr == 'name':
            attr = '_name'
        del self.__dict__[attr]
        
def main():
    p1 = Person("Rodger")
    print(p1.name)
    
    p1.name = 'Rafa'
    print(p1.name)
    del(p1.name)
    p1.abc = 'rzd'
    
    sys.exit
    
main()    
        
```

```python
# This program will demonstrate use of __getattr__ and __setattr__
# for computed attributes!


import sys

class AttrSquare:
    def __init__(self, start):             # Triggers __setattr__ !
        self.value = start              
            
    def __getattr__(self, attr):           # On undefined attr fetch
        if attr == 'x':
            return self.value ** 2
        else:
            raise AttributeError(attr)
            
    def __setattr__(self, attr, value):      # On all assignments
        if attr == 'x':
            attr = 'value'
        self.__dict__[attr] = value
        
    
def main():
    A = AttrSquare(3)             # 2 instances of class with overloading
    B = AttrSquare(32)            # Each has different state information
    print(A.x)                    # 3 ** 2
    A.x = 4
    print(A.x)                    # 4 ** 2
    print(B.x)                    # 32 ** 2 (1024)
    
    sys.exit
    
main()
            
```
